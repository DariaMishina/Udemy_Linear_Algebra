# Vector-vector multiplication: the dot product
---


```python
# many ways to compute the dot product

v1 = np.array([1, 2, 3, 4, 5])
v2 = np.array([0, -4,  -3, 6, 5])

# method 1
dp1 = sum(np.multiply(v1, v2))

# method 2
dp2 = np.dot( v1,v2 )

# method 3
dp3 = np.matmul( v1,v2 )

# method 4
dp4 = 0  # initialize

# loop over elements
for i in range(len(v1)):
    
    # multiply corresponding element and sum
    dp4 = dp4 + v1[i]*v2[i]


print(dp1,dp2,dp3,dp4)
```

# Vector-Dot product properties: associative and distributive
---


```python
## Distributive property

# create random vectors
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# the two results
res1 = np.dot( a , (b+c) )
res2 = np.dot(a,b) + np.dot(a,c)

# compare them
print([ res1,res2 ])

```

```python
## Associative property

# create random vectors
n = 5
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# the two results
res1 = np.dot( a , np.dot(b,c) )
res2 = np.dot( np.dot(a,b) , c )

# compare them
print(res1)
print(res2)


### special cases where associative property works!
# 1) one vector is the zeros vector
# 2) a==b==c
```

```python
res1 = (b@c)*a
res2 = (a@b)*c
print(res1)
print(res2)
```

```python
### special cases where associative property works!
# 1) one vector is the zeros vector
# 2) a==b==c

a = np.array([3,5,5,6])
b = np.array([3,3,5,6])
c = np.array([3,5,5,6])

res1 = (b@c)*a
res2 = (a@b)*c
print(res1)
print(res2)
```

# Vector-Vector length
---


```python
# a vector
v1 = np.array([ 1, 2, 3, 4, 5, 6 ])

# methods 1-4, just like with the regular dot product, e.g.:
vl1 = np.sqrt(sum(np.multiply(v1,v1)))

# method 5: take the norm
vl2 = np.linalg.norm(v1)

vl3 = (np.dot(v1,v1))**0.5


print(vl1,vl2,vl3)
```

The dot product from a geometric perspective
---


```python
# two vectors
v1 = np.array([ 2,  4, -3 ])
v2 = np.array([ 0, -3, -3 ])

# compute the angle (radians) between two vectors

ang = np.arccos(np.dot(v1,v2) / (np.linalg.norm(v1)* np.linalg.norm(v2)))

print(ang) # in radians

# draw them
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'b')
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'r')

plt.axis((-6, 6, -6, 6))
plt.title('Angle between vectors: %s rad.' %ang)
plt.show()

```

```python
## equivalence of algebraic and geometric dot product formulas

# two vectors
v1 = np.array([ 2,  4, -3 ])
v2 = np.array([ 0, -3, -3 ])


# algebraic
dp_a = np.dot( v1,v2 )

# geometric
dp_g = np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(ang)

# print dot product to command
print(dp_a)
print(dp_g)
```

Vector outer product
---



```python
v1 = np.array([1,2,3])
v2 = np.array ([-1,0,1])

# outer product

op1 = np.outer(v1,v2)

# looping

op2 = np.zeros((len(v1),len(v1)))
for i in range(0,len(v2)):
    for j in range (0,len(v2)):
        op2[i,j] = v1[i]*v2[j]
print(op1)
print()
print(op2)
```

```python
# set S (SET)
S1 = np.array([1, 1, 0])
S2 = np.array([1, 7, 0])

# vectors v and w
v = np.array([1, 2, 0])
w = np.array([3, 2, 1])

# draw vectors
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, S1[0]],[0, S1[1]],[.1, S1[2]+.1],'r',linewidth=3)
ax.plot([0, S2[0]],[0, S2[1]],[.1, S2[2]+.1],'r',linewidth=3)

ax.plot([0, v[0]],[0, v[1]],[.1, v[2]+.1],'g',linewidth=3)
ax.plot([0, w[0]],[0, w[1]],[0, w[2]],'b')

# now draw plane
xx, yy = np.meshgrid(range(-15,16), range(-15,16))
cp = np.cross(S1,S2)
z1 = (-cp[0]*xx - cp[1]*yy)*1./cp[2]
ax.plot_surface(xx,yy,z1)

plt.show()
```
