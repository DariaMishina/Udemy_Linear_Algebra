# Vector-vector multiplication: the dot product
---


```python
# many ways to compute the dot product

v1 = np.array([1, 2, 3, 4, 5])
v2 = np.array([0, -4,  -3, 6, 5])

# method 1
dp1 = sum(np.multiply(v1, v2))

# method 2
dp2 = np.dot( v1,v2 )

# method 3
dp3 = np.matmul( v1,v2 )

# method 4
dp4 = 0  # initialize

# loop over elements
for i in range(len(v1)):
    
    # multiply corresponding element and sum
    dp4 = dp4 + v1[i]*v2[i]


print(dp1,dp2,dp3,dp4)
```

# Vector-Dot product properties: associative and distributive
---


```python
## Distributive property

# create random vectors
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# the two results
res1 = np.dot( a , (b+c) )
res2 = np.dot(a,b) + np.dot(a,c)

# compare them
print([ res1,res2 ])

```

```python
## Associative property

# create random vectors
n = 5
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# the two results
res1 = np.dot( a , np.dot(b,c) )
res2 = np.dot( np.dot(a,b) , c )

# compare them
print(res1)
print(res2)


### special cases where associative property works!
# 1) one vector is the zeros vector
# 2) a==b==c
```

```python
res1 = (b@c)*a
res2 = (a@b)*c
print(res1)
print(res2)
```

```python
### special cases where associative property works!
# 1) one vector is the zeros vector
# 2) a==b==c

a = np.array([3,5,5,6])
b = np.array([3,3,5,6])
c = np.array([3,5,5,6])

res1 = (b@c)*a
res2 = (a@b)*c
print(res1)
print(res2)
```
